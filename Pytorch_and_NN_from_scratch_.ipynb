# Develop nueral network in pytorch from scan

```
import torch
import matplotlib.pyplot as plt
% matplotlib  inline
```

```
# create input
x= torch.tensor([[1,0,1,0],[1,0,1,1],[0,1,0,1]],dtype=torch.float)
x,x.shape
```

```
#  convert input into matrix
x=x.t()
x,x.shape
```

```
#output shape
y=torch.tensor([[1],[1],[0]],dtype=torch.float)
y=y.t()
y,y.shape
```

# Network architecture

```
inputlayer_nurons= x.shape[0]
hiddenlayer_nurons=3
outputlayer_nurones= 1 
```

```
# initializing the weight matrices
w_ih=torch.randn(inputlayer_nurons,hiddenlayer_nurons)
w_ho=torch.randn(hiddenlayer_nurons,outputlayer_nurones)
```

```
w_ih.shape,w_ho.shape
```

```
# define sigmoid function
def sigmoid(s):
  return 1/(1+torch.exp(-s))
```

```
# hidden layer activation
hiddenlayer_input=torch.matmul(w_ih.t(),x)
hiddenlayer_activations=sigmoid(hiddenlayer_input)
```

```
# get the output
outputlayer_input= torch.matmul(w_ho.t(),hiddenlayer_activations)
output=sigmoid(outputlayer_input)
```

```
output
```

```
#calculate error
error=torch.pow(y-output,2)/2
```

```
err_wrt_output=-(y-output)
output_wrt_Z2=torch.mul(output,(1-output))
z2_wrt_w_ho= hiddenlayer_activations
err_wrt_who=torch.matmul(z2_wrt_w_ho,(err_wrt_output*output_wrt_Z2).t())
```

```
err_wrt_output=-(y-output)
output_wrt_Z2=torch.mul(output,(1-output))
z2_wrt_w_h1= w_ho
h1_wrt_z1=torch.mul(hiddenlayer_activations,(1-hiddenlayer_activations))
z1_wrt_h1=x
err_wrt_wh1=torch.matmul(z1_wrt_h1,(h1_wrt_z1*torch.matmul(z2_wrt_w_h1,(output_wrt_Z2*err_wrt_output))).t())
```

```
lr=0.01
```

```
w_ho,w_ih
```

```
w_ho=w_ho-(lr*err_wrt_who)
w_ho
```

```
w_ih=w_ih-(lr*err_wrt_wh1)
w_ih
```

```
epoch=10000
```

```
error_epoch=[]
for i in range(epoch):
  hiddenlayer_input=torch.matmul(w_ih.t(),x)
  hiddenlayer_activations=sigmoid(hiddenlayer_input)
  outputlayer_input= torch.matmul(w_ho.t(),hiddenlayer_activations)
  output=sigmoid(outputlayer_input)
  error=torch.pow(y-output,2)/2

  err_wrt_output=-(y-output)
  output_wrt_Z2=torch.mul(output,(1-output))
  z2_wrt_w_ho= hiddenlayer_activations
  err_wrt_who=torch.matmul(z2_wrt_w_ho,(err_wrt_output*output_wrt_Z2).t())   
  err_wrt_output=-(y-output)
  output_wrt_Z2=torch.mul(output,(1-output))
  z2_wrt_w_h1= w_ho
  h1_wrt_z1=torch.mul(hiddenlayer_activations,(1-hiddenlayer_activations))
  z1_wrt_h1=x
  err_wrt_wh1=torch.matmul(z1_wrt_h1,(h1_wrt_z1*torch.matmul(z2_wrt_w_h1,(output_wrt_Z2*err_wrt_output))).t())
  w_ho=w_ho-(lr*err_wrt_who)
  w_ih=w_ih-(lr*err_wrt_wh1)
  error_epoch.append(torch.mean(error))
```
